```python
cat_groups = data.groupby('category')['CV'].apply(list).reset_index()
cat_groups
```

```python
groups = data.groupby('category')
random = groups.get_group('random')['CV'].to_list()
tree = groups.get_group('tree')['CV'].to_list()
mlst = groups.get_group('mlst')['CV'].to_list()
bio = groups.get_group('bioproject')['CV'].to_list()
```

```python
mann_df = pd.DataFrame(columns=['Comparison', 'p value'])
mann_df
```

```python
statistic, p_value = stats.mannwhitneyu(x=random, y=tree, alternative='greater')
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
mann_df.loc[len(mann_df)] = ['random - tree', "{:.2e}".format(p_value)]
mann_df
```

```python
statistic, p_value = stats.mannwhitneyu(x=random, y=mlst, alternative='greater')
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
mann_df.loc[len(mann_df)] = ['random - mlst', "{:.2e}".format(p_value)]
mann_df
```

```python
statistic, p_value = stats.mannwhitneyu(x=random, y=bio, alternative='greater')
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
mann_df.loc[len(mann_df)] = ['random - bioproject', "{:.2e}".format(p_value)]
mann_df
```

```python
mann_df.to_csv("mann_grouping_9.csv", index=False)
```

```python
isolate_groups = data.groupby(['Clinical/Enviromental'])['# of pseudogenes'].apply(list).reset_index()
isolate_groups
```

```python
groups = data.groupby(['Clinical/Enviromental'])
clinical = groups.get_group('clinical')['# of pseudogenes'].to_list()
enviromental = groups.get_group('environmental/other')['# of pseudogenes'].to_list()
print(clinical)
print(enviromental)
```

```python
statistic, p_value = stats.mannwhitneyu(x=clinical, y=enviromental, alternative='greater')
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
```

```python
input_list = isolate_groups['# of pseudogenes'].to_list()
```

```python
statistic, p_value = stats.mannwhitneyu(*input_list)
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
```

```python
statistic, p_value = stats.f_oneway(*input_list)
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
```

```python
results = sp.posthoc_dunn(data, val_col='# of pseudogenes', group_col='Level', p_adjust='bonferroni')
results
```

```python
results = results.reset_index()
results
```

```python
results.to_csv("level_results.csv", index=False)
```
