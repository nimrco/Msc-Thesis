```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/content/supplementary_table_with_counters.csv")
data
```

```python
data.insert(data.columns.get_loc('# pseudo_only_single') + 1, '# pseudo_only_single_normal', 0.0)
data.insert(data.columns.get_loc('# pseudo_only_other') + 1, '# pseudo_only_other_normal', 0.0)
data.insert(data.columns.get_loc('# mixed_single') + 1, '# mixed_single_normal', 0.0)
data.insert(data.columns.get_loc('# mixed_other') + 1, '# mixed_other_normal', 0.0)
data
```

```python
for index, row in data.iterrows():
  pseudo = row['# of pseudogenes']
  if pseudo == 0:
    continue
  pseudo_single = row['# pseudo_only_single'] / pseudo
  pseudo_other = row['# pseudo_only_other'] / pseudo
  mixed_single = row['# mixed_single'] / pseudo
  mixed_other = row['# mixed_other'] / pseudo
  data.at[index, '# pseudo_only_single_normal'] = pseudo_single
  data.at[index, '# pseudo_only_other_normal'] = pseudo_other
  data.at[index, '# mixed_single_normal'] = mixed_single
  data.at[index, '# mixed_other_normal'] = mixed_other
data
```

```python
data['Scaffolds'].mean()
```

```python
data.to_csv("pseudo_normal.csv", index=False)
```

```python
single = data.sort_values(by=['# pseudo_only_single_normal'], ascending=False)
single
```

```python
sns.set_theme(style='darkgrid')
```

```python
pseudogenes = data[['# pseudo_only_single_normal', '# mixed_single_normal', '# pseudo_only_other_normal', '# mixed_other_normal']]
pseudogenes
```

```python
sns.catplot(data=pseudogenes, height=10)
plt.xlabel('Pseudogene class')
plt.ylabel('# of pseudogenes')
plt.title('Distribution of pseudogenes by classification')
# plt.savefig('pseudogene_class.png', dpi=300)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.boxplot(data=pseudogenes, ax=ax)
plt.xlabel('Pseudogene class')
plt.ylabel('# of pseudogenes')
plt.title('Distribution of pseudogenes by classification')
# sns.stripplot(data=pseudogenes, ax=ax, color=".26", size=4)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.swarmplot(data=pseudogenes, ax=ax)
plt.xlabel('Pseudogene class')
plt.ylabel('# of pseudogenes')
plt.title('Distribution of pseudogenes by classification')
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.boxenplot(data=pseudogenes, ax=ax)
plt.xlabel('Pseudogene class')
plt.ylabel('# of pseudogenes')
plt.title('Distribution of pseudogenes by classification')
```
