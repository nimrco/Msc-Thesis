```python
tips = pd.read_csv('/content/tips.csv')
tips
```

```python
window_size = 7
input_list = []
```

```python
for index, _ in tips.iterrows():
  last_strain = index + window_size
  if last_strain >= 4698:
      last_strain = 4698
  strains = tips.iloc[index:last_strain]['x'].to_list()
  genes = []
  for strain in strains:
    genes.append(int(supp_df.at[strain, '# of genes']))
  input_list.append(genes)
  if last_strain == 4698:
    break
input_list
```

```python
statistic, p_value = stats.levene(*input_list)
print("statistic: {s} , p-value: {p}".format(s=statistic, p=p_value))
```
